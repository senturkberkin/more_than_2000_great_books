```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url = 'https://thegreatestbooks.org/?page='
```

```python
##TAKES A MINUTE TO LOAD###
titles= []
authors =[]

for page in range(1,55):

    html_text = requests.get(url + str(page)).text
    soup = BeautifulSoup(html_text, 'lxml')
    container = soup.find('div', class_='container')
    books = container.find_all('h4')
    for book in books:
        links = book.find_all('a')
        title =links[0].text
        
        titles.append(title)
        try:
            author = links[1].text
            
            authors.append(author)
        except:
            author = 'Unkown'
            
            authors.append(author)

```

```python
print(len(authors))

print(len(titles))
```

```python
export_data = pd.DataFrame(data={'Book Title': titles, 'Author': authors })
export_data
```

```python
final = export_data.iloc[1: , :]
final.to_csv("lots-of-great-books.csv")
```
